# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
data['Buyer Gender'].value_counts().plot(kind='bar',color='red')
plt.title('No.of Male vs Female buyers',fontsize=15)
plt.xlabel('Gender',fontsize=12)
plt.ylabel('No.of buyers',fontsize=12)
```

A bar graph is plot comparing the no.of female and male genders participated in sales

# 4. Find the top 5 cars based on their sales price.

```python
Top5_Make = data.sort_values('Sale Price',ascending=False).head(5)
Top5_Make
```

Top 5 cars data is displayed above based on sales price

```python
Top5_Make[['Make','Model']]
```

These are the 5 makes of cars that have highest sales price
